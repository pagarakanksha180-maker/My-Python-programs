Packing and Unpacking

1. Packing : The process of assigning values to a tuple.

2. Unpacking : The process of extracting values from a tuple and assigning them to variables.

```python
a=(1,2,3)  #Packing
x,y,z=a    #Unpacking
```

```python
print(a)
print(x,y,z)
```

```python
#Unpacking example

student=("Akanksha",18,"Computer")
name,age,branch=student
print("Name :",name)
print("age:",age)
print("Branch:",branch)
```

```python
numbers=(10,20,30,40,50) #Packing numbers
print(numbers)
a,b,c,d=(10,20,30,40)    #Unpacking numbers
print(a)
print(b)
print(c)
print(d)
```

```python
data=(1,2,3,4,5)  #Using * for unpacking
a,*b=data
print(a)
print(b)
```

```python
data=(100,200,300,400,500)  #Unpacking with First and last Element
first,*middle,last=data
print("First:",first)
print("Middle:",middle)
print("Last:",last)
```

```python
a=10
b=20
a,b=b,a  # Swapping Two Variables (Unpacking)
print("a=",a)
print("b=",b)
```

```python
def display (*numbers):   #Packing functional arguments
  print(numbers)
display(10,20,30,40)
```

```python
def add(a,b,c):  #Unpacking while calling a Function
  print(a+b+c)
values=(5,10,15)
add(*values)
```
