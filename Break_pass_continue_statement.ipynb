The Break , pass and continue this are the looping contro statement in python.

1.Break statement : Break immediately terminates the loop when a specified condition is met.

2.continue statement : Continue skips the current iteration and moves to the next iteration of the loop.

3.pass statement : pass is a null statement. It does nothing and is used as a placeholder where code will be added later.

1. Program to continuously take input from user.stop accepting inputs as soon as user gives input as 5

```python
while True :
  num=int(input("Enter a number :"))
  if num == 5:
    break
```

2.Take 2 numbers as input Find GCD of given 2 numbers.

```python
num1=int(input("Enter first number :"))
num2=int(input("Enter second number :"))
if num1 > num2:
  minimum=num2
else:
  minimum=num1
while minimum > 0:
  if num1 % minimum == 0 and num2 % minimum == 0:
    break
  else :
    minimum -= 1
print(f"GCD of {num1} and {num2} is",minimum)
```

3.Take 2 numbers as input Find LCM of given 2 numbers.

```python
num1=int(input("Enter first number :"))
num2=int(input("Enter second number :"))
if num1 > num2:
  maximum=num1
else:
  maximum=num2
while True:
  if maximum % num1 == 0 and maximum % num2 == 0:
    break
  maximum += 1
print(f"LCM of {num1} and {num2} is ",maximum)
```

4. Code to skip vowels in a string

```python
text=input("Enter a string :")
for ch in text :
  if ch.lower () in "aeiou":
    continue
  print(ch,end=" ")
```
