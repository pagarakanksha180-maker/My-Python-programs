Inbuilt Functions

1.minimum

```python
minimum=a[-1]
for i in a:
  if i<minimum:
    minimum=i
print(f"Lowest:",minimum)
```

2.maximum

```python
maximum=a[0]
for i in a:
  if i>maximum:
    maximum=i
print(f"highest:",maximum)
```

3.total

```python
total=0
for element in a:
  total+=element
print(f"total:",total)
```

4.count

```python
count=0
for element in a:
  count+=1
print(f"count :",count)
```

5.average

```python
avg=total/count
print(f"Average:",avg)
```

Find the sum of elements present at odd positions

```python
list=[1,3,2,4,5,6,7,9]
sum=0
n=len(list)
for i in range(0,n,2):   # for odd positions we use 0 as a start point of a for loop
  sum+=list[i]
print(f"{sum}")
```

Find the sum of elements present at odd indexes

```python
list=[1,3,2,4,5,6,7,9]
sum=0
n=len(list)
for i in range(1,n,2):        # for odd indixes we use 1 as a start point of a for loop
  sum+=list[i]
print(f"{sum}")
```

print common items in both lists

```python
List1=['JAVA','PYTHON','PHP',20,'A']
print("List1=",List1)
List2=[10,20,30,40,50]
print("List2 =",List2)
co=[]
for item in List1:
  if item in List2 and item not in co:
    co.append(item)
    print(f"common item in both lists :",co)
print(f"All common items :",co)
```

Lists Updation



1.pop()


```python
a1=[23,45,21,22,20,46,78,90]
a1.pop(3)            # 3 is a index position pop always removes the element as per the index position
print(f"after pop:",a1)
```

2.remove ()

```python
a1.remove(45)     # 45 is a element in a list remove() always remove a element as per value
print(f"After remove:",a1)
```

Difference between pop() and remove()



*   pop() removes an element using its index and returns that element
*   remove() removes the specified value from the list and does not return    anything



3.reverse()

```python
a1.reverse()   # reverse the list
print(f"After reverse:",a1)
```

4.sort()

```python
a1.sort()
print(f"After sorting:",a1)
```

5.index()

```python
index=a1.index(21) # index() always works on values
print(f"Index of 4 :",index)
```

```python
a2[0:] # here end value is not mentioned so by default it takes end value is n
```

```python
a2[0:5] # here start from 0 and end to n-1
```

```python
a2[1:3] #here start with 1 and end to n-1
```

```python
a2[:8] #here start value is not mentioned so by default it takes start value as 0 as end value is n-1
```

```python
a2[::1] #here start value and end value both are not mentioned and 1 is a step
```

```python
a2[::-1] # here step is -1 it prints list in a reverse order
```
