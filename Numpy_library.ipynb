Multi indexing:

```python
a=[[10,20,30],[40,50,60],[70,80,90]]   #multi indexing at 2D list
a[0][1]
```

```python
import numpy as np
arr=np.array([[10,20,30],[40,50,60],[70,80,90]])   #multi index
print(arr[0,1])
print(arr[1,2])
print(arr[2,0])
```

```python
import numpy as np
arr=np.array([[10,20,30],[40,50,60],[70,80,90]])   # to access a entire row
print(arr[1])
```

```python
import numpy as np
arr=np.array([[10,20,30],[40,50,60],[70,80,90]])  # to access a entire column
print(arr[:,1])
```

```python
import numpy as np
arr=np.array([[10,20,30],[40,50,60],[70,80,90]])  # negative indexing
print(arr[-1,-1])
```
