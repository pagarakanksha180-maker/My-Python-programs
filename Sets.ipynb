Set Operations

```python
s={1,2,3,4,5,6,7,8,9}
print(s)
```

```python
s.pop()   # Removes first element in the set
print(s)
```

```python
s.add(10) # Add at the end
print(s)
```

```python
s.remove(3)
print(s)
```

The main difference between pop() and remove () is :

1. pop() always removes first element in the set.

2. remove() removes the specific element in the set.

```python
s.update((1,2,9,8))
print(s)
```

Set Operations:

1. Union(A|B) : it prints All elements From A and B remove duplicates.

2. Intersection(A and B) /(A & B) : All common elements from A and B .

3. Difference (A-B)/(A or B): All elements that are in A but not in B.

4. Symmetric Difference (A^B) : All un-common elements from A and B.

```python
A={1,2,3,4,5,6,7}
B={4,5,6,7,8,9,0}
print(f"Set A : {A} \nSet B : {B}")
```

1. Union(A|B):


```python
print(A|B)
```

2. Intersection(A & B):

```python
print(A & B)
```

3. Difference(A-B):

```python
print(A-B)
```

4. Symmetric Difference (A^B):

```python
print(A^B)
```

Set operations is also done by its functions.
1. union()
2. intersection()
3. difference()
4. symmetric difference

```python
print(A.union(B))
print(A.intersection(B))
print(A.difference(B))
print(A.symmetric_difference(B))
```

```python
 # Set difference
print(A.difference(B))
print(A-B)
print((A or B)-(A and B))   # Output of these six ways to find difference is same
print(B.difference(A))
print(B-A)
print((B or A)-(B and A))
```
