4. Check whether a number is prime or not

```python
n=int(input("Enter a number :"))
count=0
for i in range(1,n+1):
  if n % i==0:
    count+=1
if count==2:
  print(f"prime number")
else:
  print(f"Not a prime number")
```

6.Find HCF of two numbers

```python
a=int(input("Enter a first number :"))
b=int(input("Enter a second number :"))
if a<b:
  small=a
else:
  small=b
hcf=1
for i in range(1,small+1):
  if a % i==0 and b % i==0:
    hcf=i
print(f"HCF : {hcf}")
```

6.to reverse a string

```python
s=input("Enter a string :")
rev=""
for i in range(len(s)-1,-1,-1):
  rev += s[i]
print(f"Reverse: {rev}")
```

7. To check whether a number is palindrome or not

```python
n=int(input("Enter a number :"))
temp=n
rev=0
for i in range(len(str(n))):
  digit=n%10
  rev=rev*10+digit
  n=n//10
if temp == rev:
  print("Palindrome")
else:
  print("Not a palindrome")
```

8. Write a program to take any 2 numbers as input from user and find LCM of two numbers.

```python
a=int(input("Enter first number :"))
b=int(input("Enter second number :"))
if a>b:
  large=a
else:
  large=b
for i in range(large,a*b+1):
    if i % a == 0 and i % b == 0:
      lcm = i
      break
print(f"The LCM of {a} and {b} is {lcm}")
```
