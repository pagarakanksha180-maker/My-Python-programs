Inbuilt Functions of Python:

1. ord():

```python
ch='A'
print(ord(ch))
```

2. chr():

```python
print(chr(65))
```

3. lower():

```python
s="HELLO"
print(s.lower())
```

4. upper():

```python
s="hello"
print(s.upper())
```

5. count():

```python
s="banana"
print(s.count('a'))
```

6. append():

```python
lst=[10,20,30]
lst.append(40) # adds element at the last
print(lst)
```

7. extend():

```python
lst=[10,20,30]
lst.extend([40,50,60])  # adds a block of elements
print(lst)
```

8. sum():

```python
lst=[10,20,30]
print(sum(lst))
```

9. len():

```python
s="Python"
print(len(s))
```

10. id():

```python
x=100
print(id(x))  #it gives address of this character
```

11. islower():

```python
s="python"
print(s.islower())   #it checks whether the string is lower or not
```

12. isupper():

```python
s="PYTHON"
print(s.isupper())  # It checks whether the string is upper or not
```

13. isalpha():

```python
s="PYTHON"
print(s.isalpha())  #It checks whether the all characters in a string is alphabets or not
```

14. isdigit():

```python
s="12345"
print(s.isdigit()) #It checks whether the all digits in a number is digits or not
```

15. index():

```python
s="Python"
print(s.index("t"))
```

16. del():

```python
lst=[10,20,30,40]
del lst[1]  #It delete element present at index 1
print(lst)
```

17. pop():

```python
lst=[10,20,30,40,50]
lst.pop()  #pop() always remove last element in the list
print(lst)
```

18. min():

```python
lst=[5,2,7,9]
print(min(lst))
```

19. max():

```python
lst=[5,3,6,9,10]
print(max(lst))
```

20. reverse():

```python
lst=[10,20,30,40,50,60]
lst.reverse()
print(lst)
```

21. replace():

```python
s=" I Like Java"
print(s.replace("Java","Python"))
```

22. find():

```python
s="Python"
print(s.find("t"))  #prints index position of t
```

23. join():

```python
words=["Welcome","To","World","Of","Python"]
print(" ".join(words))
```

24. split():

```python
s="Welcome to world of Python"
print(s.split())
```

25. type():

```python
x=10
print(type(x))
```

26. isspace():

```python
s="  "
print(s.isspace())
```

27. insert():

```python
lst=[10,20,30,40]
lst.insert(2,50)   #it adds element at specific index
print(lst)
```

28. get():

```python
student={"name":"Akanksha","age":18}
print(student.get("name"))
```

29. abs():

```python
num=-25
print(abs(num))
```

30. round():

```python
num=12.5678
print(round(num,2))
```

31. pow():

```python
print(pow(2,5))
```

32. divmod():

```python
print(divmod(10,3))
```

33. int():

```python
x="50"
print(int(x))  #It convert to integer
```

34. float():

```python
x=50
print(float(x)) # it converts to float
```

35. str():

```python
x=100
print(str(x))
```

36. bool():

```python
print(bool(1))
print(bool(0))
```

37. sorted():

```python
numbers=[7,6,3,9,2]
print(sorted(numbers))
```

38. bin():

```python
print(bin(10))
```

39. oct():

```python
print(oct(10))
```

40. hex():

```python
print(hex(10))
```
