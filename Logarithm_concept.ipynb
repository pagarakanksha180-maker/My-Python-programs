Logarithm : A logarithm is the power(exponent) to which a base must be raised to obtain a given number.

Logarithm base 2

```python
num=int(input("Enter a number :"))
count=0
while num>1:
  num=num//2
  count=count+1
print("log base 2=",count)
```
