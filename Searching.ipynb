Searching : Searching in python means finding a element exists in a collection (such as list , tuple , set or dictionary)

Two types of search:

1. Linear search
2. Binary search

1. Linear Search

Algorithm of Linear Search:

Linear Search works on both list sorted or unsorted

Step 1: Start

Step 2: Create a list of elements.

Step 3: Input the element to be searched or already define the element in the program.

Step 4 : Compare the search element with each element in the list one by one until the element is not found.

Step 5: If a match is found after checking display the element index.

Step 6: Stop

```python
a=[5,3,4,6,8,10,11,23,32]
key=int(input("Enter the key :"))  # Here key is user defined
for i in range(len(a)):
  if a[i]==key:
    print("element found at index :",i)
```

2. Binary Search

Algorithm of Binary Search:

Binary Search works only on a Sorted list.

Step 1: Start

Step 2: Input the sorted list of elements.

Step 3: Input the element to be searched (key) or already define the element in the program.

Step 4: Set low=0 and high=n-1

Step 5:  Repeat while low<=high

         Calculate mid=(low+high)//2

         If list[mid] ==key , display Index of element and stop

         If key < list[mid] , set high=mid-1

         Otherwise set low=mid+1

Step 6: If the element is not found after the loop display Element not found

Step 7: Stop



```python
# Binary Search Program

numbers=[10,20,30,40,50,60,70,80,90]
key=int(input("Enter the number to search :"))   # Here key is user defined
low=0
high=len(numbers)-1
found=False
while low <= high:
  mid=(low+high)//2
  if numbers[mid] == key:
    print("Element found at index",mid)
    found=True
    break
  elif key < numbers[mid]:
    high= mid-1
  else:
    low = mid+1
if not found:
  print("Element not found")
```
